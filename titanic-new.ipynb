```python
import pandas as pd
import numpy as np

data=pd.read_csv('./train.csv',delimiter=",")
data.head()

```

```python
line=data.shape[0]
Title=[]
Familly_Name=[]
First_Name=[]
Married_Woman=[]
for i in range(line):
    name=data.get_value(i,'Name')
    key=name.split(",")
    inter=key[1] # separate coma:  familly name "," Title "." First Name "(" Before wedding name or if "(" " nickname
    Familly_Name.append(key[0])
    key2= inter.split(".")
    Title.append(key2[0])
    key3=key2[1].split("(")
    First_Name.append(key3[0])
    #print(key3)
    if len(key3)>1:
        #print(key3[1][0])
        if key3[1][0]!='"':
            Married_Woman.append(1)
        else:
            Married_Woman.append(0)
    else:
        Married_Woman.append(0)
data['Title']=Title
data['First_Name']=First_Name
data['Familly_name']=Familly_Name
data['Married_Woman']=Married_Woman
data
```

```python
import matplotlib.pyplot as plt

plt.figure()
data['Age'].plot.hist(bins=5)
line=data.shape[0]
class_age=[]
classe_title=[]
Sexe=[]
embark=[]
# class 0 : x<19 ans
# class 1 : 19<= x < 35
# class 2 : 35<= x <50
# class 3 : 50<= x <65
# class 4 : 65<=x
for i in range(line):
    age=data.get_value(i,'Age')
    if (age <19 and age>=0):  
        class_age.append(0)
    elif (age<35 and age>= 19):
        class_age.append(1)
    elif (age<50 and age>=35):
        class_age.append(2)
    elif (age<65 and age>=50):
        class_age.append(3)
    elif (age>=65 and age<=100):
        class_age.append(4)
    else:
        class_age.append(-1)

for j in range(line):
    title=data.get_value(j,'Title')
    title=title.strip()
    #print(title)
    if title=='Mrs' or title=='Ms':
        classe_title.append(0)
    elif title=='Mr':
        classe_title.append(1)
    elif title=='Mlle' or title=='Miss':
        classe_title.append(2)
    elif title=='Master':
        classe_title.append(3)
    elif title =='Rev':
        classe_title.append(4)
    elif title =='Sir':
        classe_title.append(5)
    elif title=='the Countess':
        classe_title.append(6)
    elif title=='Major':
        classe_title.append(7)
    elif title=='Lady':
        classe_title.append(8)
    elif title == 'Dr':
        classe_title.append(9)
    elif title == 'Col':
        classe_title.append(10)
    elif title=='Capt':
        classe_title.append(11)
    elif title =='Jonkheer':
        classe_title.append(12)
    else:
        classe_title.append(-1)

for k in range(line):
    sex=data.get_value(k,'Sex')
    #sex=sex.strip()
    if sex=='male':
        Sexe.append(0)
    else:
        Sexe.append(1)
data['Sex']=Sexe

for l in range(line):
    emba=data.get_value(l,'Embarked')
    #emba=emba.strip()
    if emba=='C':
        embark.append(0)
    elif emba=='S':
        embark.append(1)
    else:
        embark.append(2)
data['Embarked']=embark
#Sex_mapping = {'female': 1,'male': 0}
#Embark_mapping={'C':0,'S':1, 'Q':2}
#data['Embarked']=data['Embarked'].map(Embark_mapping)
#data['Sex'] = data['Sex'].map(Sex_mapping )   
data['Age_Classe']=class_age
#Title_mapping={ 'Mrs' :0 ,'Ms':13 ,'Mr':1, 'Miss':2,'Mlle':14, 'Master': 3, 'Rev':4, 'Sir': 5,'the Countess':6, 'Major':7, 'Lady':8, 'Dr': 9, 'Col': 10,'Capt':11,'Jonkheer':12}
#data['Title_Category']=data['Title'].map(Title_mapping)
data['Title_Category']=classe_title
data
```

```python
Companion=[]

for p in range(line):
    count=0
    ticket_p=data.get_value(p,'Ticket')
    for l in range(line):
        ticket_l=data.get_value(l,'Ticket')
        if ticket_l==ticket_p:
            count=count+1
    compa=count-1 #remove the personstself
    Companion.append(compa)
    
data['Companion']=Companion
data
```

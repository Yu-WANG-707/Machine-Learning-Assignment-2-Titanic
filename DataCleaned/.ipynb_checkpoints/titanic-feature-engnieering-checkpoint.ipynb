```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_train=pd.read_csv('./train.csv',delimiter=",")     # load in training data
data_train.tail()
```

```python
data_train.shape
```

```python
data_test=pd.read_csv('./test.csv',delimiter=",")     # load in test data
data_test.insert(loc=2, column='Survived', value=-1)  # assign -1 to "Survived" for samples that need to be predicted
data_test.tail()
```

```python
data_test.shape
```

```python
frames = [data_train, data_test]                         # combine training data & test data together ----  make it easier for feature engineering
data = pd.concat(frames, sort = False)
data.shape
```

```python
data=data.fillna(np.nan)
print(data.isna().sum())  
```

```python
plt.figure()
data['Fare'].plot.hist(bins=300)
```

```python
mean = data["Fare"].mean()        # calculate the mean of "Fare"
mean
```

```python
median = data["Fare"].median()    # calculate the median of "Fare", the gragh is highly skewed, so here we replace NaN with median
median
```

```python
data["Fare"].replace(np.nan, median, inplace = True)
print(data.isna().sum())  
```

```python
#.  c/s/q
```

```python
data["Embarked"].replace(np.nan, "S", inplace = True)     # Replace the NaN with "S" for "Embarked"
print(data.isna().sum())  
```

```python
# *********************** Create new features by extracting information *****************************
```

```python
line=data.shape[0]
Title=[]
Familly_Name=[]
First_Name=[]
Married_Woman=[]
name = ' '
for i in range(line):
    name=data.iat[i,3]     # get "Name"
    key=name.split(",")
    inter=key[1]            # separate coma:  familly name "," Title "." First Name "(" Before wedding name or if "(" " nickname
    Familly_Name.append(key[0])
    key2= inter.split(".")
    Title.append(key2[0])
    key3=key2[1].split("(")
    First_Name.append(key3[0])
    #print(key3)
    if len(key3)>1:
        #print(key3[1][0])
        if key3[1][0]!='"':
            Married_Woman.append(1)
        else:
            Married_Woman.append(0)
    else:
        Married_Woman.append(0)
data['Title']=Title
data['First_Name']=First_Name
data['Familly_name']=Familly_Name
data['Married_Woman']=Married_Woman
# data
```

```python

plt.figure()
data['Age'].plot.hist(bins=5)
```

```python
line=data.shape[0]
class_age=[]
# class 0 : x<19 ans
# class 1 : 19<= x < 35
# class 2 : 35<= x <50
# class 3 : 50<= x <65
# class 4 : 65<=x
for i in range(line):
    age=data.iat[i,5]     # 'Age'
    if   age <19 and age>=0:
        class_age.append(0)
    elif (age<35 and age>= 19):
        class_age.append(1)
    elif (age<50 and age>=35):
        class_age.append(2)
    elif (age<65 and age>=50):
        class_age.append(3)
    elif (age>=65 and age<=100):
        class_age.append(4)
    else:
        class_age.append(-1)
        
data['Age_Classe']=class_age

```

```python
classe_title=[]

#Title_mapping={ 'Mrs' :0 ,'Ms':13 ,'Mr':1, 'Miss':2,'Mlle':14, 'Master': 3, 'Rev':4, 'Sir': 5,'the Countess':6, 'Major':7, 'Lady':8, 'Dr': 9, 'Col': 10,'Capt':11,'Jonkheer':12}
#data['Title_Category']=data['Title'].map(Title_mapping)

for j in range(line):
    title=data.iat[j,12]   # "Title"
    title=title.strip()
    #print(title)
    if title=='Mrs' or title=='Ms':
        classe_title.append(0)
    elif title=='Mr':
        classe_title.append(1)
    elif title=='Mlle' or title=='Miss':
        classe_title.append(2)
    elif title=='Master':
        classe_title.append(3)
    elif title =='Rev':
        classe_title.append(4)
    elif title =='Sir':
        classe_title.append(5)
    elif title=='the Countess':
        classe_title.append(6)
    elif title=='Major':
        classe_title.append(7)
    elif title=='Lady':
        classe_title.append(8)
    elif title == 'Dr':
        classe_title.append(9)
    elif title == 'Col':
        classe_title.append(10)
    elif title=='Capt':
        classe_title.append(11)
    elif title =='Jonkheer':
        classe_title.append(12)
    else:
        classe_title.append(-1)

data['Title_Category']=classe_title
data
```

```python
Sexe=[]
#data['Sex'] = data['Sex'].map(Sex_mapping )   
#Sex_mapping = {'female': 1,'male': 0}

for k in range(line):
    sex=data.iat[k,4] # "Sex"
    #sex=sex.strip()
    if sex=='male':
        Sexe.append(0)
    else:
        Sexe.append(1)
        
data['Sex']=Sexe
```

```python
embark=[]
#Embark_mapping={'C':0,'S':1, 'Q':2}
#data['Embarked']=data['Embarked'].map(Embark_mapping)

for l in range(line):
    emba=data.iat[l,11]  # "Embarked"
    #emba=emba.strip()
    if emba=='C':
        embark.append(0)
    elif emba=='S':
        embark.append(1)
    else:
        embark.append(2)
data['Embarked']=embark
data
```

```python
Companion=[]

for p in range(line):
    count=0
    ticket_p=data.iat[p,8]    # "ticket"
    for l in range(line):
        ticket_l=data.iat[l,8]
        if ticket_l==ticket_p:
            count=count+1
    compa=count-1 #remove the personstself
    Companion.append(compa)
    
data['Companion']=Companion
data
#data['Companion'].value_counts()
```

```python
print(data.isnull().sum()) 
```

```python
from sklearn import cross_validation
Data_ForAgeGroup=data[['Title_Category','Companion','Sex','Pclass','SibSp','Parch','Fare','Embarked','Married_Woman', 'Age_Classe']]

# data with age known
Data_ForAgeGroup_train=Data_ForAgeGroup.loc[Data_ForAgeGroup['Age_Classe']!=-1]
# separate labels and features
labels=Data_ForAgeGroup_train[['Age_Classe']]
Data_ForAgeGroup_train=Data_ForAgeGroup_train[['Title_Category','Sex','Pclass','SibSp','Parch','Fare','Embarked','Married_Woman','Companion']]

# sample to prediction
Data_ForAgeGroup_predict=Data_ForAgeGroup.loc[Data_ForAgeGroup['Age_Classe']==-1]
Data_ForAgeGroup_predict=Data_ForAgeGroup_predict[['Pclass']]
```

```python
# LogisticRegression
# crosse - validation
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(Data_ForAgeGroup_train, labels)   # multi-classification
# clf.score(Xtest, ytest)   # prediction accuracy
scores = cross_val_score(clf, Data_ForAgeGroup_train, labels, cv=10)
scores
```

```python
# Support Vector Machine
from sklearn.svm import SVC
clf = SVC(gamma='auto').fit(Data_ForAgeGroup_train, labels)     # C = 1 by default
scores = cross_val_score(clf, Data_ForAgeGroup_train, labels, cv=5)
scores
```

```python
#  Decision Tree 
from sklearn import tree
clf = tree.DecisionTreeClassifier().fit(Data_ForAgeGroup_train, labels)     
scores = cross_val_score(clf, Data_ForAgeGroup_train, labels, cv=5)
scores
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30),activation='tanh',max_iter=10000)
mlp.fit(Data_ForAgeGroup_train, labels)
scores = cross_val_score(mlp, Data_ForAgeGroup_train, labels, cv=5)
predict=mlp.predict(Data_ForAgeGroup_train)
predict
scores
#print(np.mean(scores))
```
